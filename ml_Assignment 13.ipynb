# Naive Bayes Classifier Algorithm

```python
import pandas as pd
df = pd.read_csv('spam.csv')
df.head()
```

```python
df.groupby('Category').describe()
```

```python
df['spam'] = df['Category'].apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:2]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count, y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch football game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count = v.transform(X_test)
model.score(X_test_count, y_test)
```

### using Sklearn Pipeline

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```

```python
clf.predict(emails)
```

```python
'''
Use wine dataset from sklearn.datasets to classify wines into 3 categories. 
Load the dataset and split it into test and train. 
After that train the model using Gaussian and Multinominal classifier and post 
which model performs better. Use the trained model to perform some predictions on test data.
'''
```

```python
from sklearn.datasets import load_wine
wine = load_wine()
```

```python
dir(wine)
```

```python
wine.target_names
```

```python
df = pd.DataFrame(wine.data, columns = wine.feature_names)
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df['target'] = wine.target
df.head()
```

```python
X = df.drop('target', axis = 1)
y = df.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB
g_model = GaussianNB()
m_model = MultinomialNB()
```

```python
# Gaussian Model Score
g_model.fit(X_train, y_train)
g_model.score(X_test, y_test)
```

```python
# Multinomial Model Score
m_model.fit(X_train, y_train)
m_model.score(X_test, y_test)
```

### Using cross_val_score

```python
from sklearn.model_selection import cross_val_score

# Gaussian Model score
cross_val_score(g_model, X, y, cv = 5).mean()
```

```python
# Multinomial Model score
cross_val_score(m_model, X, y, cv = 5).mean()
```

### CONCLUSION:
#### Gaussian Model has better score compared to Multinomial Model.
