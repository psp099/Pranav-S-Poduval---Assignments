# DECISION TREE ALGORITHM

```python
import pandas as pd
```

```python
df = pd.read_csv("salaries.csv")
df.head()
```

```python
inputs = df.drop('salary_more_then_100k', axis = 'columns')
target = df['salary_more_then_100k']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_company.fit_transform(inputs['job'])
inputs['degree_n'] = le_company.fit_transform(inputs['degree'])
inputs.head()
```

```python
inputs_n = inputs.drop(['company', 'job', 'degree'], axis = 1)
```

```python
inputs_n
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
model.fit(inputs_n, target)
```

```python
model.score(inputs_n, target)
```

```python
model.predict([[2, 2, 1]])
```

## EXERCISE

```python
# Predict the survival rate based on titanic dataset.
```

```python
data = pd.read_csv("titanic.csv")
data.head()
```

```python
# features: class, sex, age, fare
# target: survived
```

```python
features = data[['Pclass', 'Sex', 'Age', 'Fare']].copy()
target = data['Survived']
```

```python
import math

# Encoding
le_Sex = LabelEncoder()
features['sex_n'] = le_Sex.fit_transform(data['Sex'])

# resolve NaN's
features['Age'] = features['Age'].fillna(math.floor(features['Age'].median()))
features.head()
```

```python
features = features.drop(columns = ['Sex'])
features
```

```python
model = tree.DecisionTreeClassifier()
model.fit(features, target)
```

```python
model.predict([[1, 35, 53.1, 0]])
```

```python
model.score(features, target)
```
