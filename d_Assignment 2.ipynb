## Cleaning the data

```python
data.isnull().sum()
```

```python
student = data.drop(["Race/Ethnicity","Parental Level of Education"], axis=1) 
```

```python
student.fillna('Unknown', inplace=True)
```

```python
student.head()
```

## Relationship Analysis

```python
corelation = student.corr(numeric_only=True)
```

```python
sns.heatmap(corelation, xticklabels = corelation.columns, yticklabels = corelation.columns,
            annot = True, cmap = "coolwarm")
```

```python
sns.pairplot(student)
```

```python
sns.relplot(x = "Math Score", y = "Reading Score", hue = "Gender", data = student)
```

```python
sns.histplot(student['Math Score'])
```

```python
sns.histplot(student['Writing Score'], bins = 20)
```

```python
sns.catplot(x = 'Writing Score', kind = 'box', data = student)
```
