# Questions

### Q1. Find 10 cheapest food items suitable for a hypertensive patient (less sodium, more potassium & more magnesium).

```python
data.describe()[['Sodium', 'Potassium', 'Magnesium']]
```

```python
#considering thresholds.

sodium_threshold = 10
potassium_threshold = 700
magnesium_threshold = 100
filtered = data[
    (data['Sodium'] <= sodium_threshold) &
    (data['Potassium'] >= potassium_threshold) &
    (data['Magnesium'] >= magnesium_threshold)
]

top_10_items = filtered.sort_values(by = "Price").reset_index().head(10)

top_10_items[['food', 'Price','Sodium', 'Potassium', 'Magnesium']]
#top cheapest food for hypertensive patient. 
```

### Q2. Find 10 cheapest food items suitable for a diabetic patient (less sugars, less carbo-hydrate) with lowest Glycemic Index (GI).

```python
data.describe()[['Sugars', 'Carbohydrates','Glycemic Index']]
```

```python
sugar_threshold = 5
carb_threshold = 30
gi_threshold = 55

filtered = data[(data['Sugars']<=sugar_threshold) &
                (data['Carbohydrates']<=carb_threshold) &
                (data['Glycemic Index']<=gi_threshold)]

top_10_items = filtered.sort_values(by = "Price").reset_index().head(10)

top_10_items[['food', 'Price','Sugars', 'Carbohydrates', 'Glycemic Index']]
#top cheapest food for hypertensive patient.
```

### Q3. Find 7 costly food items with maximum nutrition value but minimum carbohydrates.

```python
data.describe()[['Price', 'Nutrition Density', 'Carbohydrates']]
```

```python
nutrition_threshold = 100
carb_threshold = 20

filtered = data[(data['Nutrition Density']>=nutrition_threshold) &
                (data['Carbohydrates']<=carb_threshold)]

top_7_nutritious_lowcarbs = filtered.sort_values(by = 'Price', ascending = False).reset_index().head(7)
top_7_nutritious_lowcarbs[['food', 'Price', 'Nutrition Density', 'Carbohydrates']]
```

### Q4. Find 10 cheapest food items suitable for a diabetic osteoporosis patient (less sugar, max. Calcium, max. Phosphorous, Max. Vitamin D)

```python
data.describe()[['Sugars', 'Calcium', 'Phosphorus', 'Vitamin D']]
```

```python
# Thresholds
sugar_max = 2             
calcium_min = 0     
phosphorus_min = 0
vitamin_d_min = 2        


filtered = data[
    (data['Sugars'] <= sugar_max) &
    (data['Calcium'] >= calcium_min) &
    (data['Phosphorus'] >= phosphorus_min) &
    (data['Vitamin D'] >= vitamin_d_min)]

top10_diabetic_osteoporosis = filtered.sort_values(by='Price', ascending=True).reset_index().head(10)

top10_diabetic_osteoporosis[['food', 'Price', 'Sugars', 'Calcium', 'Phosphorus', 'Vitamin D']]
```

### Q6. Find costliest food items which are rich in all types of vitamins.

```python
data.columns
```

```python
data.describe()[['Vitamin A', 'Vitamin B1', 'Vitamin B11', 'Vitamin B12', 'Vitamin B2',
       'Vitamin B3', 'Vitamin B5', 'Vitamin B6', 'Vitamin C', 'Vitamin D',
       'Vitamin E', 'Vitamin K']]
```

```python
vitamin_columns = ['Vitamin A', 'Vitamin B1', 'Vitamin B11', 'Vitamin B12', 'Vitamin B2',
       'Vitamin B3', 'Vitamin B5', 'Vitamin B6', 'Vitamin C', 'Vitamin D',
       'Vitamin E', 'Vitamin K']
vitamin_thresholds = data[vitamin_columns].quantile(0.20)
vitamin_thresholds
```

```python
rich_vitamins = data.copy()

for vitamin in vitamin_columns:
    rich_vitamins = rich_vitamins[rich_vitamins[vitamin] >= vitamin_thresholds[vitamin]]

costly_vitamins = rich_vitamins.sort_values(by= "Price", ascending = False).reset_index()
costly_vitamins[['food', 'Price', 'Vitamin A', 'Vitamin B1', 'Vitamin B11', 'Vitamin B12', 'Vitamin B2',
       'Vitamin B3', 'Vitamin B5', 'Vitamin B6', 'Vitamin C', 'Vitamin D',
       'Vitamin E', 'Vitamin K']].head()
```

### Q15. 15 food items suitable for keto diet - maximum protein, min carbohydrates, maximum fat

```python
# High protein: top 25%
protein_thresh = data['Protein'].quantile(0.75)

# Low carb: bottom 25%
carb_thresh = data['Carbohydrates'].quantile(0.25)

# High fat: top 25%
fat_thresh = data['Fat'].quantile(0.75) 
```

```python
keto_friendly = data[
    (data['Protein'] >= protein_thresh) &
    (data['Carbohydrates'] <= carb_thresh) &
    (data['Fat'] >= fat_thresh)]

top15_keto = keto_friendly.sort_values(by='Protein', ascending=False).head(15).reset_index()
top15_keto[['food', 'Protein', 'Carbohydrates', 'Fat']]
```
