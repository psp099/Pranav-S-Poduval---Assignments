# LINEAR REGRESSION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
plt.xlabel("area(sqft)")
plt.ylabel("price(US)")
plt.scatter(df.area, df.price, color = 'red', marker = '+')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#plotting linear regression line along with scatter plot
plt.xlabel("area", fontsize = 20)
plt.ylabel("price", fontsize = 20)
plt.scatter(df.area, df.price, color = 'red', marker = '+')
plt.plot(df.area, reg.predict(df[['area']]), color = 'blue')
```

```python
d = pd.read_csv("areas.csv")
d
```

```python
d['prices'] = reg.predict(d)
d
```

```python
d.to_csv("prediction.csv", index = False)
```

```python
pd.read_csv("prediction.csv")
```

## Exercise Question

```python
data = pd.read_csv("canada_per_capita_income.csv")
data.columns
```

```python
model = linear_model.LinearRegression()
model.fit(data[['year']], data['per capita income (US$)'])
```

```python
#comparison plot
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(data['year'], data['per capita income (US$)'], color = 'red', marker = '+')
plt.plot(data['year'], model.predict(data[['year']]))
```

```python
#PREDICTION IN 2020
p = model.predict([[2020]])
print("Per capita income in the year 2020:", p)
```
