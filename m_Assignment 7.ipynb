# LOGISTIC REGRESSION - MULTI CLASS

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
# sample images of first five numbers of digit dataset.
plt.gray() 
for i in range(5):
    plt.matshow(digits.images[i]) 
```

```python
dir(digits)    #to see all the attributes and methods that an object has.
```

```python
digits.data[0]    # to see the matrix values in specified digit's image.
```

```python
digits.target  # the digits each image represents
```

## Create and train logistic regression model.

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2)
```

```python
model.fit(X_train, y_train)
```

### Measure accuracy of our model

```python
model.score(X_test, y_test)
```

```python
model.predict(digits.data[0:5])
```

### Confusion Matrix

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
# PLOT THE CONFUSION MATRIC OF y_test v/s y_predicted 
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
